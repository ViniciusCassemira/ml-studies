## Usando valores ficticios para cálculo - matriz confusão

```python
# Definindo variávels
VP = 20
VN = 4
FP = 3
FN = 2
```

#### Funções de cálculo

```python
def precisao(vp, fp):
    return vp / (vp + fp)

def revocacao(vp, fn):
    return vp / (vp + fn)

def acuracia(vp, vn, fp, fn):
    return (vp + vn) / (vp + vn + fp + fn)

def f_measure(precisao, revocacao):
    return 2 * ((precisao * revocacao) / (precisao + revocacao))
```

#### Usando variáveis


```python
precisao_calculada = precisao(VP, FP) * 100
revocacao_calculada = revocacao(VP, FN) * 100
acuracia_calculada = acuracia(VP, VN, FP, FN) * 100
f_measure_calculada = f_measure(precisao_calculada, revocacao_calculada)


print(f"Precisão: {precisao_calculada:.2f}%")
print(f"Revocação: {revocacao_calculada:.2f}%")
print(f"Acurácia: {acuracia_calculada:.2f}%")
print(f"F measure: {f_measure_calculada:.2f}%")
```
